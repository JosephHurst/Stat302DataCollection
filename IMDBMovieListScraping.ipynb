```python
%matplotlib inline
import matplotlib.pyplot as plt
from urllib.request import urlopen
from bs4 import BeautifulSoup
from dateutil.parser import parse
from datetime import datetime
import json
import csv
import pandas as pd
import re
import numpy as np
```

```python
def getAllMovies(url):
    bsobj = BeautifulSoup(urlopen(url), 'lxml')
    movies = []
    movies += getMovies(bsobj)
    while len(movies) < 9954:
        nextUrl = getNextPage(bsobj)
        bsobj = BeautifulSoup(urlopen(nextUrl), 'lxml')
        movies += getMovies(bsobj)
        print(nextUrl)
    return movies
```

```python
def getMovies(bsobj):
    results = bsobj.find('div', {'class': 'list compact'}).findAll('td', {'class': 'title'})
    return [(obj.getText(), obj.find('a').get('href')) for obj in results]
```

```python
def getNextPage(bsobj):
    baseUrl = 'http://www.imdb.com/list/ls057823854/'
    return baseUrl + bsobj.find('div', {'class': 'pagination'}).findAll('a')[-1].get('href')
```

```python
url = 'http://www.imdb.com/list/ls057823854/?start=1&view=compact&sort=listorian:asc&defaults=1&scb=0.5257108344826827'
movieNames = getAllMovies(url)
```

```python
def getMovieName(index):
    return movieNames[index - 1][0]
def getMovieUrl(index):
    return 'http://www.imdb.com'+movieNames[index - 1][1]
```

```python
random = np.unique([np.random.randint(9954) for _ in range(1000)])
```
